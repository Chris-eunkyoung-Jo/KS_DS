```
import json #for more information, https://www.json.org/json-en.html
import requests

# To get data from all pages, Make a user defined funtion to reuse the same code.
def get_page_data(url, show_repos = False):
  data = requests.get(url)
  repos = json.loads(data.text)
  print("cur page:", url[-6:])
  page_data = []
  for i, each in enumerate(repos): 
    if each != 'storeInfos':
      print(i, each, repos[each])
    else:
      first_store = lambda dic_list: ( dic_list[0] )
      if len(repos[each])>0 :
        print(i, each, first_store(repos[each]))
        page_data = repos[each]
  return page_data
```

**Draw a pie chart with labels and legend**

```
#https://matplotlib.org/3.1.0/gallery/pie_and_polar_charts/pie_and_donut_labels.html
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ["375 g flour",   "75 g sugar",  "250 g butter", "300 g berries"]
data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data), textprops=dict(color="w"))
ax.legend(wedges, ingredients, title="Ingredients", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=8, weight="bold")
ax.set_title("Matplotlib bakery: A pie")
plt.show()
```
